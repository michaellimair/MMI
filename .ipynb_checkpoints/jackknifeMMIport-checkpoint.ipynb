```python
#Import Python libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

#R to Python interfacing
import rpy2.robjects as robj
import rpy2.robjects.packages as rpackages
import rpy2.robjects.numpy2ri
rpy2.robjects.numpy2ri.activate()
from rpy2.robjects.numpy2ri import numpy2ri

#use seaborn engine to display matplotlib graph
%matplotlib inline
sns.set()

#define random seed
iter = 1000
np.random.seed(iter)
    
#import jackknife MI estimator library
#requires R 3.4.4 and the jackknife library in the computer
mJMI = rpackages.importr('mJMI')
```

```python
def get_model(model, n, coef):
    #define noise with coef parameter
    if (model <= 16):
        noise = pow(10,(1/24*(coef-1)))
    elif ((model > 17 and model <= 24) or (model >= 26)):
        n1 = 76 - 3*coef
    else:
        n1 = 104 - 4*coef
    
    #define models
    #model 1: linear with symmetric additive noise
    if (model == 1):
        xs = np.random.normal(size=n)
        ys = (2/3)*xs + noise*np.random.normal(size=n)
    
    #model 2: linear with asymmetric additive noise
    if (model == 2):
        xs = np.random.normal(size=n)
        ys = (0.05)*xs + noise*((pow(np.random.normal(size=n),2)) - 1)
    
    #model 3: quadratic with additive noise
    if (model == 3):
        xs = np.random.normal(size=n)
        ys = (2/3)*xs*xs + noise*np.random.normal(size=n)
    
    #model 4: circle with additive noise
    if (model == 4):
        thetas = 2*math.pi*np.random.uniform(size=n)
        xs = 10*np.cos(thetas) + noise*np.random.normal(size=n)
        ys = 10*np.sin(thetas) + noise*np.random.normal(size=n)
    
    #model 5: spiral circle with additive noise
    if (model == 5):
        u = np.random.uniform(size=n)
        xs = 3*u*np.sin(u*math.pi) + noise*np.random.normal(size=n)
        ys = 3*u*np.cos(u*math.pi) + noise*np.random.normal(size=n)
    
    #model 6: cloud with additive noise
    if (model == 6):
        z1 = np.random.uniform(size=n)
        z2 = np.random.uniform(size=n)
        xcs = np.floor(5*z1)
        ycs = 2*np.floor(2*z2) + (xcs%2)
        xs = 10*(xcs + np.random.uniform(n)) + noise*np.random.normal(size=n)
        ys = 10*(ycs + np.random.uniform(n)) + noise*np.random.normal(size=n)
    
    #model 7: sin with additive noise
    if (model == 7):
        xs = np.random.uniform(size=n)
        ys = np.sin((20+7*noise)*math.pi*xs)
    
    #model 8: diamond with additive noise
    if (model == 8):
        tempX = np.random.uniform(size=n)*2 - 1
        tempY = np.random.uniform(size=n)*2 - 1
        theta = -math.pi/4
        
        rr = np.array([[np.cos(theta), np.sin(theta)],
                       [-np.sin(theta), np.cos(theta)]])
        tempmat = np.array([tempX, tempY]).transpose()
        temp = np.matmul(tempmat, rr)
        xs = 10*temp[:,0] + noise*np.random.normal(size=n)
        ys = 10*temp[:,1] + noise*np.random.normal(size=n)
        
    #model 9: x-para with additive noise
    if (model == 9):
        xs = np.random.uniform(size=n)*2 - 1
        ys = (4*xs*xs) * (pow((np.random.uniform(size=n)>0.5), 2)-1) + noise*np.random.normal(size=n)
   
    #model 10: step with additive noise
    if (model == 10):
        xs = np.random.uniform(size=n)*8 - 2
        err = np.random.uniform(size=n)
                       
        xerr = xs + noise*err
        xerrorder = np.sort(xerr)
        
        ys = (np.logical_and(xerr > xerrorder[80], xerr < xerrorder[241]))*1
    
    #model 11: conditional variance
    if (model == 11):
        xs = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
            #generate a matrix with size 5*n and shape n rows, 5 columns
        e = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
        
        ys = np.sqrt(0.5*xs*xs + noise)*e
    
    #model 12: multivariate logarithm with additive noise
    if (model == 12):
        xs = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
        e = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
        ys = np.log(xs*xs) + 2*noise*e
    
    #model 13: multivariate polynomial with additive noise1
    if (model == 13):
        xs = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
        e = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
        ys1 = pow(xs[:,0:2] + 4*(xs[:,0:2]), 2) + 5*noise*e[:,0:2]
        ys2 = 5*noise*e[:,2:5]
        ys = np.c_[ys1, ys2]
    
    #model 14: multivariate polynomial with additive noise2
    if (model == 14):
        xs = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
        e = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
        ys1 = 3*(xs[:,0:2]) + 2.5*pow(xs[:,0:2], 2) + 5*noise*e[:,0:2]
        ys2 = 5*noise*e[:,2:5]
        ys = np.c_[ys1, ys2]
    
    #model 15: multivariate logarithm with additive noise
    if (model == 15):
        xs = np.array(np.random.normal(loc=0, scale=1, size=20*n).reshape(n, 20))
        e = np.array(np.random.normal(loc=0, scale=1, size=20*n).reshape(n, 20))
        ys = np.log(xs*xs) + 2*noise*e
    
    #model 16: multivariate linear with additive noise
    #original code has a bug, size of xs and ys matrix is not the same
    if (model == 16):
        return
#         xs = np.array(np.random.normal(loc=0, scale=1, size=50*n).reshape(n, 50))
#         consmt = np.zeros((50,5))
#         repeatarr = np.repeat(1,10)
#         consmt[0:10,0] = repeatarr
#         consmt[10:20,1] = repeatarr
#         consmt[20:30,2] = repeatarr
#         consmt[30:40,3] = repeatarr
#         consmt[40:50,4] = repeatarr
#         e = np.array(np.random.normal(loc=0, scale=1, size=5*n).reshape(n, 5))
#         ys = np.matmul(0.3*xs, consmt) + noise*e
    
    #model 17: linear with contaminated noise
    if (model == 17):
        x1 = np.random.uniform(low=-3, high=3, size=n-n1)
        y1 = np.random.uniform(low=-3, high=3, size=n-n1)
    
        xs = np.random.normal(size=n1)
        ys = (2/3)*xs
    
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]
        
    #model 18: quadratic with contaminated noise
    if (model == 18):
        x1 = np.random.uniform(low=-3, high=3, size=n-n1)
        y1 = np.random.uniform(low=-1, high=7, size=n-n1)
    
        xs = np.random.normal(size=n1)
        ys = (2/3)*pow(xs, 2)
    
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]
    
    #model 19: circle with contaminated noise
    if (model == 19):
        x1 = np.random.uniform(low=-12, high=12, size=n-n1)
        y1 = np.random.uniform(low=-12, high=12, size=n-n1)
    
        thetas = 2*math.pi*np.random.uniform(size=n1)
        xs = 10*np.cos(thetas)
        ys = 10*np.sin(thetas)
    
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]
    
    #model 20: spiral circle with contaminated noise
    if (model == 20):
        x1 = np.random.uniform(low=-12, high=10, size=n-n1-25)
        y1 = np.random.uniform(low=-12, high=14, size=n-n1-25)
    
        u = np.random.uniform(size=n1+25)*4
        xs = 3*u*np.sin(u*math.pi)
        ys = 3*u*np.cos(u*math.pi)
    
        xs = np.r_[x1, xs]
        ys = np.r_[y1, ys]

    #model 21: cloud with contaminated noise
    if (model == 21):
        x1 = np.random.uniform(low=-5, high=45, size=n-n1)
        y1 = np.random.uniform(low=-5, high=35, size=n-n1)
        z1 = np.random.uniform(size=n1)
        z2 = np.random.uniform(size=n1)

        xcs = np.floor(5*z1)
        ycs = 2*np.floor(2*z2) + (xcs % 2)
        
        xs = 10*xcs
        ys = 10*ycs
        
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]        

    #model 22: sin with contaminated noise
    if (model == 22):
        x1 = np.random.uniform(low=-math.pi, high=6*math.pi, size=n-n1)
        y1 = np.random.uniform(low=-3, high=3, size=n-n1)
    
        xs = 5*math.pi*np.random.uniform(size=n1)
        ys = 2*np.sin(xs)
    
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]

    #model 23: diamond with contaminated noise
    if (model == 23):
        x1 = np.random.uniform(low=-15, high=15, size=n*3-n1)
        y1 = np.random.uniform(low=-15, high=15, size=n*3-n1)
        
        x2 = np.random.uniform(size=n1)*2 - 1
        y2 = np.random.uniform(size=n1)*2 - 1
        theta = -math.pi/4
        
        rr = np.array([[np.cos(theta), np.sin(theta)],
                       [-np.sin(theta), np.cos(theta)]])
        tempmat = np.array([x2, y2]).transpose()
        temp = np.matmul(tempmat, rr)
        
        xs = 10*temp[:,0]
        ys = 10*temp[:,1]
        
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]        
        
    #model 24: x-para with contaminated noise
    if (model == 24):
        x1 = np.random.uniform(low=-1.5, high=1.5, size=n-n1)
        y1 = np.random.uniform(low=-5, high=5, size=n-n1)
        
        xs = np.random.uniform(size=n1)*2 - 1
        ys = (pow(xs, 2)) * ((np.random.uniform(size=n1)>0.5)*2 - 1)
            
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]
        
    #model 25: step with contaminated noise
    if (model == 25):
        x1 = np.random.uniform(low=-5, high=5, size=n-n1)
        y1 = np.random.uniform(low=-0.5, high=1.5, size=n-n1)
        xs = np.random.uniform(size=n1)*8 - 4
        xsorder = np.sort(xs)
        
        ys = (np.logical_and(xs > xsorder[int(n1/4)], xs < xsorder[int(n1/4*3+1)]))*1
        
        xs = np.r_[xs, x1]
        ys = np.r_[ys, y1]
        
    #model 26: multivariate logarithm with contaminated noise
    if (model == 26):
        x1 = np.array(np.random.uniform(low=-3, high=3, size=5*(n-n1)).reshape(n-n1, 5))
        y1 = np.array(np.random.uniform(low=-10, high=4, size=5*(n-n1)).reshape(n-n1, 5))
        
        xs = np.array(np.random.normal(loc=0, scale=1, size=5*n1).reshape(n1, 5))
        ys = np.log(pow(xs, 2))
                    
        xs = np.concatenate((xs, x1))
        ys = np.concatenate((ys, y1))
        
    #model 27: multivariate polynomial with contaminated noise1
    if (model == 27):
        x1 = np.array(np.random.uniform(low=-2, high=2, size=5*(n-n1)).reshape(n-n1, 5))
        y1 = np.array(np.random.uniform(low=-2, high=18, size=5*(n-n1)).reshape(n-n1, 5))
        
        xs = np.array(np.random.normal(loc=0, scale=1, size=5*n1).reshape(n1, 5))
        e = np.array(np.random.normal(loc=0, scale=1, size=5*n1).reshape(n1, 5))
        
        ys1 = (xs[:,0:2]) + 4*pow(xs[:,0:2], 2)
        ys1 = np.array(ys1.reshape(n1,2))
        ys2 = np.zeros((n1,3))
        
        ys = np.c_[ys1, ys2]
        
        xs = np.concatenate((xs, x1))
        ys = np.concatenate((ys, y1))

        
    #model 28: multivariate polynomial with contaminated noise2
    if (model == 28):
        x1 = np.array(np.random.uniform(low=-2, high=2, size=5*(n-n1)).reshape(n-n1, 5))
        y1 = np.array(np.random.uniform(low=-2, high=18, size=5*(n-n1)).reshape(n-n1, 5))
        
        xs = np.array(np.random.normal(loc=0, scale=1, size=5*n1).reshape(n1, 5))
        e = np.array(np.random.normal(loc=0, scale=1, size=5*n1).reshape(n1, 5))
        
        ys1 = 3*(xs[:,0:2]) + 2.5*pow(xs[:,0:2], 2)
        ys1 = np.array(ys1.reshape(n1,2))
        ys2 = np.zeros((n1,3))
        
        ys = np.c_[ys1, ys2]
        
        xs = np.concatenate((xs, x1))
        ys = np.concatenate((ys, y1))
    
    return xs, ys
```

```python
#Datasets
def plot_data(model, n, coef):
    xs, ys = get_model(model, n, coef)
    plt.figure(figsize=(4,4))
    plt.scatter(xs, ys, c='r')
    plt.title('Dataset for Model %s and Coefficient 1' % (model))
    #make figures smaller
    plt.show()

for i in range(1, 29):
    if (i == 16): #skip model 16
        continue
    else:
        plot_data(i, 320, 1)
```

```python
def jackknifeMI(model, n, coef):
    """
    Function to calculate mutual information with Jackknife approach
    
    Inputs:
        model: Model number as defined in function
        n: Number of samples
        coef: Parameter for noise level
        
    Function returns the mutual information
    """
    
    xs, ys = get_model(model, n, coef)
    
    #reshape dataset to become a 2d array with 1 column
    xs = xs.reshape(-1,1)
    ys = ys.reshape(-1,1)
        
    #convert data to R matrix
    xr, xc = xs.shape
    x = robj.r.matrix(xs, nrow=xr, ncol=xc)
    yr, yc = ys.shape
    y = robj.r.matrix(ys, nrow=yr, ncol=yc)
  
    #run the jackknife algorithm and return the result in function
    jmi = mJMI.mJMICpp(x, y, 1000) #1000 simulations
    return jmi.rx2('mi')[0]
```

```python
#run 28 models of dataset
model_length = 28
#coefficient denoting noise levels
coef_ran = 25
#sample size of 320
n = 320

for model in range(1, model_length + 1):
    if (model == 16): #skip model 16
        continue
    MI = [] #array containing mutual information for a given coefficient
    MI_coef = [] #array for coefficient number
    for coef in range(1, coef_ran + 1):
        MI.append(jackknifeMI(model, n, coef)) #store MI
        MI_coef.append(coef) #store coefficient
    
    #Plot individual graphs of MI vs Coefficient from different data models
    plt.scatter(MI_coef, MI, c=np.random.rand(3,))
    plt.ylabel("Mutual Information")
    plt.xlabel("Coefficient")
    plt.title('MI vs Coefficient for Model %s' % (model))
    plt.show()
```
